```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline #para activar graficos en el notebook
```

```python
x = np.linspace(-15,25,100) 
x2 = np.linspace(0,5,5)
y = x**2*np.cos(x)-3*x**3 #func trig dan valores en radianes
y2 = x2**2*np.cos(x2)-3*x2**3
```

```python
plt.plot(x,y)
```

```python
#opcion 1 para definir funciones
fx=lambda x: x**2*np.cos(x)-3*x**3 # x no tiene valor 
```

```python
fx(x2) #devuelve valor para e√± arreglo x2
```

```python
#opcion 2 para definir funciones
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3 #pueden ser arreglos, numeros o una variable nummero y otra arreglo
    return z
```

```python
fx2
```

```python
plt.plot(x,fx2(x,y))
```

```python
x3=np.random.rand(10)
fx2(x3,2)
```

```python
def recta (a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))#la posicion es (x,m,b) a menos que se le indique otra cosa
plt.plot(x,recta(x,3,-5)) # esta recta pasa por -5
plt.grid()
```

```python
def recta(a,m=1,b=0,label= 'x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
x=np.linspace(-2,2,20)
recta(x)
recta(x,-2,1,'-x+1')
plt.legend()

plt.grid()
```
